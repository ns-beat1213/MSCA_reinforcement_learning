```python
# probability distribution for the initial value.
p = [0.1, 0.5, 0.4]

# transition probability matrix
P = np.array([[0.5,0.4,0.1],[0.3,0.5,0.2],[0.2,0.3,0.5]],dtype=float)

# simulation of the Markov chain
Y = np.zeros(K,dtype=int)
S = np.array([1,2,3],dtype=float)
k = 0
for u in U:
# simulation of initial values
    if k==0:
        if u <= p[0]:
            j = 0
            Y[k] = S[j]
        elif u <= p[0]+p[1]:
            j = 1
            Y[k] = S[j]
        else:
            j = 2
            Y[k] = S[j]
# simulation of the transitions
    else:
        if u <= P[i,0]:
            j = 0
            Y[k] = S[j]
        elif u <= P[i,0]+P[i,1]:
            j = 1
            Y[k] = S[j]
        else:
            j = 2
            Y[k] = S[j]
    k = k + 1
    i = j
```

```python
# 遷移行列の初期化
transition_matrix = np.zeros((3, 3))

# 遷移数を計算
for i in range(len(Y) - 1):
    current_state = Y[i] - 1
    next_state = Y[i + 1] - 1
    transition_matrix[current_state][next_state] += 1

# 各状態からの遷移確率を計算
transition_matrix /= np.sum(transition_matrix, axis=1, keepdims=True)
```

# 4.1 n-step  transition matrix

P(n + m) = P(n)P(m)

```python
T = 10
n = 3
P = np.zeros((T,n,n),dtype=np.float64)
Pmat = np.array([[0.5,0.4,0.1],[0.3,0.5,0.2],[0.2,0.3,0.5]],dtype=float)

P[0] = Pmat

P[0] = Pmat
for t in range(1,T):
    P[t] = np.matmul(P[t-1],Pmat)
```

```python
# 9つのグラフを描画する
fig, axs = plt.subplots(3, 3, figsize=(10, 10))

# グラフを1つずつ描画する
for i, ax in enumerate(axs.flatten()):
    x = np.arange(10)
    y = P[:, i // 3, i % 3]
    ax.plot(x, y)
    ax.set_title(f"({i // 3+1}, {i % 3+1})")
    ax.set_ylim(0, 0.55)

# グラフを整形する
plt.tight_layout()

plt.show()
```
